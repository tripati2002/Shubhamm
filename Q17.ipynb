```python
n=int(input("enter total number of elements:"))
l1=[]
print("enter the elements:")
for i in range(n):
  a=int(input())
  l1.append(a)
print(l1)
n1=int(input("enter total number of elements:"))
l2=[]
print("enter the elements:")
for i in range(n1):
  a1=int(input())
  l2.append(a1)
print(l2)
l3=set(l1)-set(l2)
print( "list after subtracting elements ",list(l3))
```
